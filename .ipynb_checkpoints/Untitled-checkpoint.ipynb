```python
n=int(input("enter the number "))
class vehicle:
    def __init__(self, wheel, seat):
        self.wheel =wheel
        self.seat =seat
class truck(vehicle): 
    def __init__(self,wheel=6,seat=2): #default values are given as 6 and 2
        self.wheel=wheel
        self.seat=seat
class car(vehicle): 
    def __init__(self,wheel=4,seat=5): #default values are given as 4 and 5
        self.wheel=wheel
        self.seat=seat   
class bike(vehicle): 
    def __init__(self,wheel=2,seat=1): #default vaues are given as 2 and 1
        self.wheel=wheel
        self.seat=wheel 
        
#-----------------------------------------------------------------------------------------------        
 
def createtruck(d):
    tr=truck()
    d['truck']=d['truck']+1
    d['wheel']=d['wheel']-tr.wheel
    d['seat']=d['seat']-tr.seat
    if d['wheel']<0 or d['seat']<0:
        return 1
    else:
        return d
def createcar(d):
    cr=car()
    d['car']=d['car']+1
    d['wheel']=d['wheel']-cr.wheel
    d['seat']=d['seat']-cr.seat
    if d['wheel']<0 or d['seat']<0:
        return 1
    else:
        return d
def createbike(d):
    bk=bike()
    print(bk.wheel, bk.seat)
    d['bike']=d['bike']+1
    d['wheel']=d['wheel']-bk.wheel
    d['seat']=d['seat']-bk.seat
    if d['wheel']<0 or d['seat']<0:
        return 1
    else:
        return d
#------------------------------------------------------------------------------------------------    
        
        
        
        

d={'truck':0,'bike':0,'car':0,'wheel':n,'seat':n}
while d['wheel']>0 and d['seat']>0:
    trwh=d['wheel']//6  #gives floor value of total wheels that are required
    trs=d['seat']//2    #gives floor value of total seats that are required
    tottruck=min(trwh,trs) #total number of trucks that can be created 
   

    crwh=d['wheel']//4
    crs=d['seat']//5
    totcar=min(crwh,crs) #total number of cars that can be created
    

    bkwh=d['wheel']//2
    bks=d['seat']//1
    totbikes=min(bkwh,bks) #total number of bikes that can be created
    
    

    
    m=0
    d1={'maketruck':tottruck,'makebike':totbikes,',makecar':totcar}
    for key,val in d1.items():
        if val>m:
            m=val
            maxim=key
    
           
        
    if maxim=='maketruck':
        print("t",d)
        createtruck(d)
        
    elif maxim=='makebike':
        print("b",d)
        createbike(d)
        
    elif maxim=='makecar':
        print("c")
        createcar(d)
      
    


print(d.values())    

```
